```python
# constraint (or boundary) function
def boundary_check(pos):
    x = pos[0][0]
    y = pos[0][1] 
    val = ((x-2.5)/30)**2+((y-2.5)/40)**2
    return val>=1

# reference direction of movements
direction_ref = np.array([[0,10],[10,0],[0,-10],[-10,0]])
# number of simulations
sims = 10000
# maximum number of time steps
n = 10000
end_times=np.ones(sims)*n

# starting simulations
for j in range(sims):
    position_new = np.zeros((1,2))
    for k in range(n):
        direction = np.random.randint(4, size=1) # get a random direction
        position_new = position_new + direction_ref[direction] # update position
        if boundary_check(position_new): # check whether boundary is crossed
            end_times[j]=k+1
            break

avg_time=np.mean(end_times)
print("Average time taken = ",avg_time)    
(unique, counts) = np.unique(end_times, return_counts=True)
frequencies = np.asarray((unique, counts/sims)).T
print(frequencies)
```
